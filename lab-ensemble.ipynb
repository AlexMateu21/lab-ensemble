```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
df = spaceship.dropna().reset_index(drop=True)
df.isnull().sum()
```

```python
df = df.drop(columns=["PassengerId","Name"], axis=1)
df
```

```python
features = df.drop("Transported", axis=1)
target = df["Transported"]
```

**Perform Train Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0, stratify=target)
```

```python
num_cols = X_train.select_dtypes(include=["number"]).columns
cat_cols = X_train.select_dtypes(include=["object", "bool"]).columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)
```

```python
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

**Model Selection** - now you will try to apply different ensemble methods in order to get a better model

- Bagging and Pasting

```python
from sklearn.ensemble import BaggingClassifier

bag = BaggingClassifier(random_state=0)
bag.fit(X_train, y_train)

bag.score(X_test, y_test)
```

- Random Forests

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)

rf.score(X_test, y_test)
```

- Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=0)
gb.fit(X_train, y_train)

gb.score(X_test, y_test)
```

- Adaptive Boosting

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(random_state=0)
ada.fit(X_train, y_train)

ada.score(X_test, y_test)
```

Which model is the best and why?

Gradient Boosting performs best here since it has the higher accuracy, but close to random forest
that is also very strong since it combines multiple decision trees, therefore reducing variance
